```python
def CreateDate(InpString):
    return(pd.Timestamp(datetime.datetime.strptime(InpString,'%Y-%m')))

def getQuarter(new_date):
    return(new_date.quarter)

# Defining a function to get month
def getMonth(new_date):
    return(new_date.month)
 
# Defining a function to get year
def getYear(new_date):
    return(new_date.year)
```

```python
 def read(file_name):
   name=file_name+'.csv'
   file_name=pd.read_csv(name) 
   file_name.iloc[:,0]=file_name.iloc[:,0].astype('string')
   file_name.iloc[:,0]=file_name.iloc[:,0].str.strip() 
   file_name.iloc[:,0]=file_name.iloc[:,0].str.replace(' ','-')
   file_name['new_date']=file_name.iloc[:,0].apply(CreateDate)
   file_name['month']=file_name['new_date'].apply(getMonth)
   file_name['year']=file_name['new_date'].apply(getYear)
   file_name['quarter']=file_name['new_date'].apply(getQuarter)
   indexes=file_name[file_name['year']<2016].index
   file_name.drop(indexes,axis=0,inplace=True)  
   return(file_name)
```
